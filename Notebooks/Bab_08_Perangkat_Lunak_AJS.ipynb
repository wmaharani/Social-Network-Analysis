## 8.2.2 NetworkX — Ragam Pembangunan Graf

Reproduksi seluruh jenis graf yang disebutkan di buku Bagian 8.2.2.

```python
G_simple = nx.Graph()
G_directed = nx.DiGraph()
G_multigraph = nx.MultiGraph()
G_multidigraph = nx.MultiDiGraph()

G_weighted = nx.Graph()
G_weighted.add_edge("A", "B", weight=5)

print("Jenis graf yang tersedia di NetworkX:")
print(" - nx.Graph()        : graf sederhana (tidak berarah)")
print(" - nx.DiGraph()       : graf berarah")
print(" - nx.MultiGraph()    : multigraf (boleh banyak sisi antar 2 simpul)")
print(" - nx.MultiDiGraph()  : multigraf berarah")
print(f" - Graf berbobot: G.add_edge(u, v, weight=...) -> {list(G_weighted.edges(data=True))}")
```

## 8.2.2 Perhitungan Metrik Jaringan (ringkasan fungsi)

Semua fungsi metrik yang dibahas pada Bab 4–6, dijalankan sekaligus di sini sebagai referensi cepat.

```python
G_ref = nx.karate_club_graph()

print("Degree centrality      :", dict(list(nx.degree_centrality(G_ref).items())[:3]), "...")
print("Betweenness centrality :", dict(list(nx.betweenness_centrality(G_ref).items())[:3]), "...")
print("Closeness centrality   :", dict(list(nx.closeness_centrality(G_ref).items())[:3]), "...")
print("Eigenvector centrality :", dict(list(nx.eigenvector_centrality(G_ref).items())[:3]), "...")
print("Clustering per simpul  :", dict(list(nx.clustering(G_ref).items())[:3]), "...")
print("Average clustering     :", round(nx.average_clustering(G_ref), 4))
```

## 8.2.2 Deteksi Komunitas & Ekspor Data

Ekspor graf ke format yang kompatibel dengan Gephi (GraphML, GML) — alur kerja NetworkX → Gephi.

```python
# Deteksi komunitas
komunitas = nx.community.louvain_communities(G_ref, seed=42)
print(f"Jumlah komunitas terdeteksi: {len(komunitas)}")

# Simpan hasil sebagai atribut simpul agar ikut terekspor
komunitas_map = {}
for i, c in enumerate(komunitas):
    for node in c:
        komunitas_map[node] = i
nx.set_node_attributes(G_ref, komunitas_map, "komunitas")

betweenness = nx.betweenness_centrality(G_ref)
nx.set_node_attributes(G_ref, betweenness, "betweenness")

# Ekspor ke berbagai format
nx.write_graphml(G_ref, "jaringan_karate.graphml")
nx.write_gml(G_ref, "jaringan_karate.gml")
nx.write_adjlist(G_ref, "jaringan_karate.adjlist")

print("\nFile berhasil diekspor: jaringan_karate.graphml, .gml, .adjlist")
print("File .graphml dapat langsung dibuka di Gephi untuk eksplorasi visual lebih lanjut.")
```
